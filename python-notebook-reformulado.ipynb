**Problema:**

## Aula 3


Distribuição de Moedas Digitais

Um grupo de amigos criou um jogo online e decidiram distribuir moedas digitais de forma justa entre eles. Marina, a criadora do jogo, sugeriu a seguinte regra:

Cada amigo, exceto Marina, deve receber exatamente o mesmo número de moedas. Marina deve receber o triplo da quantidade que cada amigo recebe. Todos concordaram, pois a divisão era exata, sem sobrar nem faltar moedas.

**Problema:**

Crie um programa que determine a quantidade de moedas que Marina recebeu, dado o número total de moedas disponíveis e o número de amigos de Marina.

**Entrada:**

* A primeira linha contém um número inteiro `total_moedas`, que representa o número de moedas digitais disponíveis.
* A segunda linha contém um número inteiro `numero_amigos`, que representa a quantidade de amigos de Marina (excluindo a própria Marina).

```python
total_moedas = int(input("Digite o total de moedas: "))
numero_amigos = int(input("Digite o número de amigos: "))

moedas_amigos = total_moedas // (numero_amigos + 3)
moedas_marina = 3 * moedas_amigos

print(f"Marina recebeu {moedas_marina} moedas.")
```

### Desafio

Depois de distribuir as moedas, Marina descobriu que havia mais moedas em um baú secreto do jogo. Ela decidiu que essas moedas extras também devem ser distribuídas de forma justa entre ela e seus amigos, mantendo a mesma regra:

Cada amigo recebe uma quantidade igual de moedas.
Marina recebe o triplo da quantidade de moedas que cada amigo recebe.
Ajude Marina a distribuir todas as moedas, incluindo as extras, de maneira justa, sem deixar nenhuma moeda sobrando.

Entrada

* número total de moedas iniciais.
* número de amigos de Marina (excluindo a própria Marina).
* número de moedas extras encontradas no baú.
* calcular a nova quantidade total de moedas, somando as extras.
* dividir as moedas entre Marina e seus amigos, garantindo que Marina receba o triplo de moedas que cada amigo.
exibir a quantidade de moedas que Marina recebeu.

**Entrada**

* Um número inteiro `numero_estudantes`, representando a quantidade de estudantes na visita.
* Um número inteiro `numero_professores`, representando a quantidade de professores na visita.

```python
numero_estudantes = int(input("Digite a quantidade de estudantes: "))
numero_professores = int(input("Digite a quantidade de professores: "))
resposta_sim = "Pode subir"
resposta_nao = "Não pode subir"

if numero_estudantes + numero_professores <= 40:
  print(resposta_sim)
else:
  print(resposta_nao)
```

**Labirinto Mágico:**


Labirinto Mágico é um jogo onde uma esfera brilhante percorre um caminho até chegar ao portal final. O jogador ganha cristais dependendo do caminho que a esfera percorre. Ele pode controlar o trajeto da esfera ativando algumas alavancas no labirinto. Cada alavanca pode estar em duas posições: 0 (desativada) ou 1 (ativada).

No labirinto mostrado na imagem, há duas alavancas. A alavanca X está na posição 1 (ativada) e a alavanca Y está na posição 0 (desativada). Isso faz com que a esfera chegue no portal Verde.

O programa que você deve escrever precisa, dado o valor das posições de X e Y, informar por qual dos três portais (Azul, Verde ou Roxo) a esfera vai passar.

**Entrada**

As entradas correspondem a um valor inteiro X (0 ou 1) e um valor inteiro Y (0 ou 1) que indicam as posições das alavancas do labirinto.



![Diagrama do Labirinto Mágico](https://i.imgur.com/placeholder123.png)

*(Descrição: Diagrama mostrando um labirinto com duas alavancas X e Y, e três saídas: Azul (topo), Verde (meio) e Roxo (base). As setas indicam os caminhos baseados nas posições das alavancas.)*

```python
x = int(input("Digite a posição da alavanca X (0 ou 1):"))
y = int(input("Digite a posição da alavanca Y (0 ou 1):"))

if x==0 and y==0:
  print("Roxo")
elif x==0 and y==1:
    print("Roxo")
elif x==1 and y==1:
  print("Azul")
else:
  print("Verde")
```

**Problema:**

Dona Beatriz é mãe de três filhos com idades diferentes. Este ano, a soma das idades de seus três filhos é igual à idade dela.

Você deve escrever um programa que, dada a idade de Dona Beatriz e as idades conhecidas de dois de seus filhos, determine a idade do terceiro filho e identifique quem é o mais velho.

**Entrada**

* Um número inteiro `idade_beatriz` representando a idade de Dona Beatriz.
* Um número inteiro `idade_filho1` representando a idade de um dos filhos.
* Um número inteiro `idade_filho2` representando a idade de outro filho.

```python
def calcula_idade_maior(idade_beatriz, idade_filho1, idade_filho2):
  idade_filho3 = idade_beatriz - idade_filho1 - idade_filho2

  idade_maxima = max(idade_filho1, idade_filho2, idade_filho3)

  return idade_filho3, idade_maxima

calcula_idade_maior(75, 18, 25)
```
